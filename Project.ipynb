         Exploratory Analysis of simulated E-Commerce Transcation Data

1.python code

```python

# Exploratory Analysis of Simulated E-Commerce Transaction Data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# 1. Generate Synthetic Dataset


np.random.seed(42)
num_records = 1000

product_categories = ['Electronics', 'Clothing', 'Groceries', 'Sports', 'Beauty']
regions = ['North', 'South', 'East', 'West']

data = {
    'customerAge': np.random.randint(18, 65, num_records),
    'purchaseAmount': np.abs(np.random.normal(150, 40, num_records).round(2)),
    'productCategory': np.random.choice(product_categories, num_records),
    'region': np.random.choice(regions, num_records)
}

df = pd.DataFrame(data)

# 2. Data Cleaning & Verification


print("Dataset Preview:")
print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nData Types:")
print(df.dtypes)


# 3. Descriptive Statistics


print("\nDescriptive Statistics for Purchase Amount:")
print(df['purchaseAmount'].describe())


# 4. Group-Level Analysis


# Mean by product category
mean_by_category = df.groupby('productCategory')['purchaseAmount'].mean()
print("\nAverage Purchase Amount by Product Category:")
print(mean_by_category)

# Mean by region
mean_by_region = df.groupby('region')['purchaseAmount'].mean()
print("\nAverage Purchase Amount by Region:")
print(mean_by_region)


# 5. Visualizations


# Histogram of purchaseAmount
plt.figure(figsize=(8,5))
plt.hist(df['purchaseAmount'], bins=20)
plt.title("Histogram of Purchase Amount")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.show()

# Bar chart: average purchaseAmount per productCategory
plt.figure(figsize=(8,5))
mean_by_category.plot(kind='bar')
plt.title("Average Purchase Amount Across Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Average Purchase Amount")
plt.show()
```

2.Descriptive stastistic output

The descriptive statistics for the purchase amount variable show the minimum,maximum,mean,standard deviation,and quartile distribution.these values help the overall spending behaviour of customers in the e-commerece dataset.

Average Purchase Amount by Product Category


This table shows how customer spending varies across the five product categories.Some categories demonstrate higher average purchase amount,suggesting higher-priced items or stronger customer demand.

Average Purchase Amount by Region


The regional average reveal differences in purchase behavior across North, South,East,and West regions,indicating how geographic facrorse may influence customer spending levels.
