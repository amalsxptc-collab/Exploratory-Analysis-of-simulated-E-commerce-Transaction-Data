Exploratory Analysis of Simulated E-commerce Transaction Data

```python

# STEP 1 – IMPORT LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime, timedelta
np.random.seed(42)     # for reproducible results
```

```python

# STEP 2 – GENERATE SYNTHETIC E-COMMERCE DATA

n_rows = 10000   # at least 10,000 records as required

# Generate basic columns
customer_ids = np.random.randint(1000, 5000, size=n_rows)
ages = np.random.randint(18, 65, size=n_rows)
genders = np.random.choice(['Male', 'Female', 'Other'], size=n_rows, p=[0.48, 0.48, 0.04])

product_categories = np.random.choice(
    ['Electronics', 'Fashion', 'Home Essentials', 'Groceries', 'Sports'],
    size=n_rows,
    p=[0.25, 0.25, 0.2, 0.2, 0.1]
)

quantities = np.random.randint(1, 6, size=n_rows)

base_prices = {
    'Electronics': 15000,
    'Fashion': 2000,
    'Home Essentials': 3000,
    'Groceries': 500,
    'Sports': 4000
}
prices = [base_prices[cat] + np.random.randint(-2000, 2000) for cat in product_categories]

discounts = np.random.choice([0, 5, 10, 15, 20, 25], size=n_rows, p=[0.15, 0.2, 0.25, 0.2, 0.15, 0.05])

payment_methods = np.random.choice(
    ['UPI', 'Credit Card', 'Debit Card', 'Net Banking', 'Cash on Delivery'],
    size=n_rows,
    p=[0.4, 0.25, 0.15, 0.1, 0.1]
)

regions = np.random.choice(
    ['Tier-1', 'Tier-2', 'Tier-3'],
    size=n_rows,
    p=[0.5, 0.3, 0.2]
)

# Dates in the last 12 months
start_date = datetime.today() - timedelta(days=365)
order_dates = [start_date + timedelta(days=int(x)) for x in np.random.randint(0, 365, size=n_rows)]

# Build DataFrame
df = pd.DataFrame({
    "CustomerID": customer_ids,
    "Age": ages,
    "Gender": genders,
    "ProductCategory": product_categories,
    "Quantity": quantities,
    "UnitPrice": prices,
    "DiscountPercent": discounts,
    "PaymentMethod": payment_methods,
    "Region": regions,
    "OrderDate": order_dates
})

# Add FinalAmount after discount
df["GrossAmount"] = df["Quantity"] * df["UnitPrice"]
df["DiscountAmount"] = df["GrossAmount"] * df["DiscountPercent"] / 100
df["FinalAmount"] = df["GrossAmount"] - df["DiscountAmount"]

# Add unique TransactionID (IMPORTANT – this was missing earlier)
df["TransactionID"] = range(1, len(df) + 1)

# Reorder columns to start with TransactionID
cols = ["TransactionID"] + [c for c in df.columns if c != "TransactionID"]
df = df[cols]

df.head()
```

```python

# STEP 3 – BASIC DATA QUALITY CHECKS AND CLEANING

print("Shape of dataset:", df.shape)
print("\nColumn info:")
print(df.info())

print("\nMissing values per column:")
print(df.isna().sum())

# Drop duplicate TransactionIDs just in case (should be none)
df = df.drop_duplicates(subset=["TransactionID"])

# Example: remove clearly negative or impossible values if any (safety check)
df = df[df["FinalAmount"] > 0]

print("\nShape after cleaning:", df.shape)
```

```python

# STEP 4 – CREATE HELPER COLUMNS (AGE GROUP, MONTH, etc.)

# Age Group
bins = [18, 25, 35, 45, 60, 80]
labels = ['18-25', '26-35', '36-45', '46-60', '60+']
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels, right=True)

# Month-Year for trend analysis
df["OrderMonth"] = df["OrderDate"].dt.to_period("M").astype(str)

df.head()
```

```python

# STEP 5 – DESCRIPTIVE STATISTICS (Deliverable 2)

numeric_cols = ["Age", "Quantity", "UnitPrice", "GrossAmount", "DiscountAmount", "FinalAmount"]
desc_stats = df[numeric_cols].describe().T
desc_stats
```

```python

# INTERPRETATION TEXT FOR DESCRIPTIVE STATISTICS (copy this as your Deliverable 2 text block)

descriptive_text = """
The descriptive statistics provide an overall view of the synthetic e-commerce dataset.
The average customer age is around the mid-thirties, suggesting that most buyers are
working adults. Quantity values show that customers usually purchase between one and
three items per order, with only a few large-quantity transactions. The mean unit price
and the wide range between minimum and maximum values confirm that the catalogue
contains both low-cost items such as groceries and high-value electronics.

The GrossAmount and FinalAmount fields show that the typical transaction value lies in
a moderate range, while the maximum values represent occasional high-ticket purchases.
A relatively high standard deviation for GrossAmount and FinalAmount indicates that
spending behaviour differs greatly between customers. DiscountAmount also has a wide
spread, which reflects promotional activity with different discount levels across orders.
These statistics are helpful for detecting outliers, understanding central tendencies, and
identifying how varied the customer purchase patterns are in this simulated dataset.
"""
print(descriptive_text)
```

```python

# STEP 6 – GROUP-BY ANALYSIS (Deliverable 3)

# Revenue by Product Category
category_summary = df.groupby("ProductCategory")["FinalAmount"].agg(["count", "sum", "mean"]).reset_index()
category_summary.rename(columns={"count": "NumTransactions",
                                 "sum": "TotalRevenue",
                                 "mean": "AvgOrderValue"}, inplace=True)

# Revenue by Region
region_summary = df.groupby("Region")["FinalAmount"].agg(["count", "sum", "mean"]).reset_index()
region_summary.rename(columns={"count": "NumTransactions",
                               "sum": "TotalRevenue",
                               "mean": "AvgOrderValue"}, inplace=True)

# Revenue by Payment Method
payment_summary = df.groupby("PaymentMethod")["FinalAmount"].agg(["count", "sum", "mean"]).reset_index()
payment_summary.rename(columns={"count": "NumTransactions",
                                "sum": "TotalRevenue",
                                "mean": "AvgOrderValue"}, inplace=True)

# Revenue by Age Group
agegroup_summary = df.groupby("AgeGroup")["FinalAmount"].agg(["count", "sum", "mean"]).reset_index()
agegroup_summary.rename(columns={"count": "NumTransactions",
                                 "sum": "TotalRevenue",
                                 "mean": "AvgOrderValue"}, inplace=True)

category_summary, region_summary, payment_summary, agegroup_summary
```

```python

# DETAILED INTERPRETATION FOR GROUP-BY ANALYSIS (Deliverable 3)

groupby_text = """
The group-by analysis helps identify which customer segments and categories contribute
most to revenue. Product-category results show that Electronics generates the highest
total revenue and also has the largest average order value, confirming that it is the key
driver of sales in this dataset. Fashion and Home Essentials have lower average order
values but a higher number of transactions, which indicates frequent but smaller
purchases. Groceries contribute many low-value orders, reflecting everyday buying habits.

Regional analysis reveals that Tier-1 cities account for the majority of transactions and
revenue, suggesting higher purchasing power and a stronger online shopping culture in
these locations. Tier-2 and Tier-3 regions still contribute meaningfully, but with smaller
average order values. The payment-method summary shows that UPI and Credit Card are
the most preferred options, both in transaction count and revenue, indicating customer
trust in digital payments. Age-group analysis highlights that customers between 26–45
years generate the highest revenue, aligning with the profile of mid-career professionals
with stable income. Overall, these grouped insights can guide marketing strategies,
regional promotions, and payment-partner tie-ups.
"""
print(groupby_text)
```

```python

# STEP 7 – VISUALIZATIONS (Deliverable 4)

sns.set(style="whitegrid")

# 1. Revenue by Product Category
plt.figure(figsize=(8,5))
sns.barplot(data=category_summary.sort_values("TotalRevenue", ascending=False),
            x="ProductCategory", y="TotalRevenue")
plt.title("Total Revenue by Product Category")
plt.xticks(rotation=30)
plt.show()

# 2. Revenue by Region
plt.figure(figsize=(6,4))
sns.barplot(data=region_summary.sort_values("TotalRevenue", ascending=False),
            x="Region", y="TotalRevenue")
plt.title("Total Revenue by Region")
plt.show()

# 3. Transactions by Payment Method
plt.figure(figsize=(8,5))
sns.barplot(data=payment_summary.sort_values("NumTransactions", ascending=False),
            x="PaymentMethod", y="NumTransactions")
plt.title("Number of Transactions by Payment Method")
plt.xticks(rotation=30)
plt.show()

# 4. Monthly Revenue Trend
monthly_revenue = df.groupby("OrderMonth")["FinalAmount"].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_revenue, x="OrderMonth", y="FinalAmount", marker="o")
plt.title("Monthly Revenue Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
